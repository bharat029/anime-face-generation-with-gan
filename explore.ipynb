```python
import os
from PIL import Image
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
res_range = [math.inf, 0]
res = []
ctr = 0

fnames = os.listdir('data/data')

for i, fname in enumerate(fnames):
    img_size = Image.open(os.path.join('data/data', fname)).size    

    if img_size[0] > res_range[1]: res_range[1] = img_size[1]
    elif img_size[0] < res_range[0]: res_range[0] = img_size[0]
    
    if img_size[0] < 75: 
        ctr += 1
        # os.remove(os.path.join('data/data', fname))
    
    res.append(img_size[0])

    print(f'\r{i}', end='')

res_range, ctr, len(fnames) - ctr
```

```python
plt.hist(res, 50)
plt.show()
```
